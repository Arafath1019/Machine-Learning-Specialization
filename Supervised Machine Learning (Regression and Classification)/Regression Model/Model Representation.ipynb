Goal: Learn to implement the model f(w,b) for Linear Regression with one variable

Notations:

    x = Training Example, feature values (x_train)

    y = Training Example, targets (y_train)

    x(i), y(i) = ith training example (x_i, y_i)

    m = number of training examples

    w = parameter: weight

    b = parameter: bias
    
    f(w,b)(x(i)) = The result of the model evaluation at x(i) parameterized by w,b
    
    f(w,b)(x(i)) = w*x(i) + b

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1.0, 2.0]) # x_train is the input/feature variable
y_train = np.array([300.0, 500.0]) # y_train is the target/labeled variable
```

```python
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
m = len(x_train) # m is the number of training examples
print(f"Number of training examples: {m}")
```

```python
# Training examples for i=0
i = 0
x_i = x_train[i]
y_i = y_train[i]
print(f"(x(i), y(i)) = ({x_i}, {y_i})")
```

```python
# Training example for i=1
i = 1
x_i = x_train[i]
y_i = y_train[i]
print(f"(x(i), y(i)) = ({x_i}, {y_i})")
```

Plotting the data:

```python
# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='g')

# Set the title
plt.title("Housing Prices")

# Set the y-axis label
plt.ylabel("Price (in 1000s of dollars)")

# Set the x-axis label
plt.xlabel("Size (1000 sqft)")

plt.show()
```

Model Function

```python
w = 100
b = 100
print(f"w : {w}")
print(f"b : {b}")
```

```python
def compute_model_outpute(x,w,b):
    """
    Compute the prediction of a linear model
    
    Args:
        x(ndarray(m,)): Data, m examples
        w,b(scalar): model perameters
    
    Returns
        y (ndarray(m,)): target values
    """
    
    m = len(x)
    f_wb = np.zeros(m)
    
    for i in range(m):
        f_wb[i] = w*x[i] + b
    
    return f_wb
```

```python
temp_f_wb = compute_model_outpute(x_train, w, b)
```

Plot our model prediction

```python
plt.plot(x_train, temp_f_wb, c='b', label='Our Model Prediction')
```

Plot the data points

```python
plt.scatter(x_train, y_train, marker='x', c='r', label='Actual Values')
plt.title('Housing Prices')
plt.ylabel('Price (in 1000s of dollars)')
plt.xlabel("Price (1000 sqft)")
plt.legend()
plt.show()
```
